```python
import torch
import torch.nn as nn
torch.set_printoptions(precision=9,sci_mode=False)
```

```python
conv2d_padding_1 = nn.Conv2d(3,6,3,1,1)
```

```python
conv2d_padding_0 = nn.Conv2d(3,6,3,1,0)
```

```python
in_tensor = torch.randn((2,3,4,4),dtype=torch.float32)
```

```python
#输入的nhwc格式
in_tensor.permute(0,2,3,1).flatten()
```

```python
padding_0_result = conv2d_padding_0(in_tensor)
```

```python
padding_1_result = conv2d_padding_1(in_tensor)
```

* 保存权重值

```python
padding_1_weight = conv2d_padding_1.weight.detach()
```

```python
padding_1_weight = padding_1_weight.permute(0,2,3,1)
```

```python
padding_1_bias = conv2d_padding_1.bias.detach()
```

```python
padding_1_bin = torch.cat([padding_1_weight.flatten(),padding_1_bias],0)
```

```python
padding_1_bin.flatten()
```

```python
padding_1_bin.numpy().tofile("./2dconv_6_3_3_3.bin")
```
