#### 将conv3d的weight和bias导出为二进制权重文件-ByteNN

```python
conv3d_weights_ndhwc = conv3d.weight.detach().permute(0,2,3,4,1).flatten()# 3d kernel shape (N,K_d,K_h,K_w,K_c) 6,3,2,2,2
```

```python
conv3d_bias = conv3d.bias.detach()
```

```python
conv3d_weights = torch.cat([conv3d_weights_ndhwc,conv3d_bias],0)
conv3d_weights.numpy().tofile("./3dconv_nopadding_6_3_2_2_2.bin")
```

```python
ndhwc_inpu = inpu.permute(0,2,3,4,1)
```

```python
ndhwc_inpu.flatten()
```

#### 将conv3d导出为onnx格式

```python
torch.onnx.export(conv3d, inpu, "conv3d.onnx", input_names=["data"], output_names=["output"], opset_version=9, verbose=True)
```

### 带padding的3D卷积

```python
p_conv3d =  nn.Conv3d(3, 6 , 2 , 2 , 1) #padding_3
```

```python
p_conv3d
```

```python
output = p_conv3d(inpu)#2, 6, 3, 3, 3 ncdhw
```

```python
output.shape
```

```python
output.permute(0,2,3,4,1).flatten() #ndhwc output
```

```python
p_conv3d_weight = p_conv3d.weight.detach().permute(0,2,3,4,1).flatten()
```

```python
p_conv3d_weight
```

```python
p_conv3d_bias = p_conv3d.bias.detach()
```

```python
p_weight = torch.cat([p_conv3d_weight,p_conv3d_bias],0)
```

```python
p_weight.numpy().tofile("./padding_3dconv.bin")
```

```python
torch.onnx.export(p_conv3d, inpu, "padding_conv3d.onnx", input_names=["data"], output_names=["output"], opset_version=9, verbose=True)
```
