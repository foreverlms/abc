时间步数为$T$ 的损失函数L定义为：

$$L = \frac{1}{T}\sum_{t=1}^T \ell(O_t,y_t)$$

$\ell$是对于单个样本的损失函数，即选定的损失函数



时间步数为3的RNN展开计算图

总的损失函数对时间步$t$的的激活值$O_t$的梯度计算(其他时刻的$\ell$对$O_t$的偏导数为0)：
$$
\frac{\partial L}{\partial O_t} = \frac{\partial L}{\partial O_t} = \frac{1}{T} \times \frac{\partial \ell(O_t,y_t)}{\partial O_t}
$$

那么$L$对于模型参数的求导需要通过链式法则(prod):

$$\frac{\partial L}{\partial W_{qh}} = \sum_{t=1}^T \mathrm{prod} \left( \frac{\partial L}{\partial O_t} , \frac{\partial O_t}{\partial W_{qh}} \right) = \sum_{t=1}^T \frac{\partial L}{\partial O_t} h_t^T
$$
