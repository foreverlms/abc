```python
import numpy as np
import tensorflow as tf

root_path = "/Users/bytedance/Downloads/train/"

x_array = np.random.rand(1, 1026)
x_array = x_array.astype(np.float32)
x_array.tofile(root_path + "x_ndarray.bin")
np.save(root_path + "x_ndarray", x_array)

w1_array = 2* np.random.random((256, 1024)) -1
w1_array = w1_array.astype(np.float32)
w1_array.tofile(root_path + "w1_ndarray.bin")
np.save(root_path + "w1_ndarray", w1_array)

w2_array = 2 * np.random.random((1,256)) - 1
w2_array = w2_array.astype(np.float32)
w2_array.tofile(root_path + "w2_ndarray.bin")
np.save(root_path + "w2_ndarray", w2_array)

x = tf.convert_to_tensor(x_array, dtype=tf.float32)
w1 = tf.convert_to_tensor(w1_array, dtype=tf.float32)

b1 = tf.ones((256))
w2 = tf.convert_to_tensor(w2_array, dtype=tf.float32)
b2 = tf.ones((1))

labels = tf.convert_to_tensor([1], dtype=tf.float32)
print("labels: ", labels)

with tf.GradientTape(persistent=True) as t:
    t.watch(x)
    t.watch(w1)
    t.watch(b1)
    t.watch(w2)
    t.watch(b2)
    slice1 = tf.slice(x, [0,0], [-1,1024])
    slice1.numpy().tofile(root_path + "slice1.bin")
    np.save(root_path + "slice1", slice1.numpy())
    
    slice2 = tf.slice(x, [0,1024], [-1, 2])
    slice2.numpy().tofile(root_path + "slice2.bin")
    np.save(root_path + "slice2", slice2.numpy())
    
    fc1_mul = tf.linalg.matmul(slice1, w1, transpose_b=True)
    fc1_mul_numpy = fc1_mul.numpy()
    fc1_mul_numpy.tofile(root_path + "fc1_mul_numpy.bin")
    np.save(root_path + "fc1_mul_numpy", fc1_mul_numpy)
    
    fc1_add = tf.add(fc1_mul, b1)
    fc1_add.numpy().tofile(root_path + "fc1_add.bin")
    np.save(root_path + "fc1_add", fc1_add.numpy())
    
    fc1_relu = tf.nn.relu(fc1_add)
    fc1_relu.numpy().tofile(root_path + "fc1_relu.bin")
    np.save(root_path + "fc1_relu", fc1_relu.numpy())
    
    fc2_mul = tf.linalg.matmul(fc1_relu, w2, transpose_b=True)
    fc2_mul.numpy().tofile(root_path + "fc2_mul.bin")
    np.save(root_path + "fc2_mul", fc2_mul.numpy())
    
    fc2_add = tf.add(fc2_mul, b2)
    fc2_add.numpy().tofile(root_path + "fc2_add.bin")
    np.save(root_path + "fc2_add", fc2_add.numpy())
    
    concat = tf.concat([fc2_add, slice2], 1)
    concat.numpy().tofile(root_path + "concat.bin")
    np.save(root_path + "concat", concat.numpy())
    
    print("concat: ", concat)
    sum1 = tf.reduce_sum(concat, axis=[1], keepdims=False)
    sum1.numpy().tofile(root_path + "sum1.bin")
    np.save(root_path + "sum1", sum1.numpy())
    
    print("sum1: ", sum1)
    loss = tf.nn.sigmoid_cross_entropy_with_logits(labels, sum1)
    loss.numpy().tofile(root_path + "loss.bin")
    np.save(root_path + "loss", loss.numpy())
    
    print("loss: ", loss)
   # bce = tf.keras.losses.BinaryCrossentropy(from_logits=True)
   # bce_value = bce(labels, sum1).numpy()
   # print("bce_value: ", bce_value)

dloss_sum = t.gradient(loss, sum1)
dloss_sum.numpy().tofile(root_path + "dloss_sum.bin")
np.save(root_path + "dloss_sum", dloss_sum.numpy())
print("dloss_sum: ", dloss_sum)

dloss_concat = t.gradient(loss, concat)
dloss_concat.numpy().tofile(root_path + "dloss_concat.bin")
np.save(root_path + "dloss_concat", dloss_concat.numpy())
print("dloss_concat: ", dloss_concat)

dloss_slice2 = t.gradient(loss, slice2)
dloss_slice2.numpy().tofile(root_path + "dloss_slice2.bin")
np.save(root_path + "dloss_slice2", dloss_slice2.numpy())
print("dloss_slice2: ", dloss_slice2)

dloss_fc2add = t.gradient(loss, fc2_add)
dloss_fc2add.numpy().tofile(root_path + "dloss_fc2add.bin")
np.save(root_path + "dloss_fc2add", dloss_fc2add.numpy())
print("dloss_fc2add: ", dloss_fc2add)

dloss_b2 = t.gradient(loss, b2)
dloss_b2.numpy().tofile(root_path + "dloss_b2.bin")
np.save(root_path + "dloss_b2", dloss_b2.numpy())
print("dloss_b2: ", dloss_b2)

dloss_w2 = t.gradient(loss, w2)
dloss_w2.numpy().tofile(root_path + "dloss_w2.bin")
np.save(root_path + "dloss_w2", dloss_w2.numpy())
print("dloss_w2: ", dloss_w2)

dloss_fc2mul = t.gradient(loss, fc2_mul)
dloss_fc2mul.numpy().tofile(root_path + "dloss_fc2mul.bin")
np.save(root_path + "dloss_fc2mul", dloss_fc2mul.numpy())
print("dloss_fc2mul: ", dloss_fc2mul)

dloss_fc1ac = t.gradient(loss, fc1_relu)
dloss_fc1ac.numpy().tofile(root_path + "dloss_fc1ac.bin")
np.save(root_path + "dloss_fc1ac", dloss_fc1ac.numpy())
print("dloss_fc1ac: ", dloss_fc1ac)

dloss_b1 = t.gradient(loss, b1)
dloss_b1.numpy().tofile(root_path + "dloss_b1.bin")
np.save(root_path + "dloss_b1", dloss_b1.numpy())
print("dloss_b1: ", dloss_b1)

dloss_w1 = t.gradient(loss, w1)
dloss_w1.numpy().tofile(root_path + "dloss_w1.bin")
np.save(root_path + "dloss_w1", dloss_w1.numpy())
print("dloss_w1: ", dloss_w1)

dloss_fc1add = t.gradient(loss, fc1_add)
dloss_fc1add.numpy().tofile(root_path + "dloss_fc1add.bin")
np.save(root_path + "dloss_fc1add", dloss_fc1add.numpy())
print("dloss_fc1add: ", dloss_fc1add)

dloss_fc1 = t.gradient(loss, fc1_mul)
dloss_fc1.numpy().tofile(root_path + "dloss_fc1.bin")
np.save(root_path + "dloss_fc1", dloss_fc1.numpy())
print("dloss_fc1: ", dloss_fc1)

dloss_slice1 = t.gradient(loss, slice1)
dloss_slice1.numpy().tofile(root_path + "dloss_slice1.bin")
np.save(root_path + "dloss_slice1", dloss_slice1.numpy())
print("dloss_slice1: ", dloss_slice1)

# Use the tape to compute the derivative of z with respect to the
# intermediate value y.
```
